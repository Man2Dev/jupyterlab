```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy.stats import ttest_ind

# Sample data
coffee = [88, 85, 90, 86, 84, 89, 92, 87, 91, 90]
no_coffee = [82, 80, 79, 81, 78, 83, 77, 84, 79, 80]

# Combine into a DataFrame
df = pd.DataFrame({
    'Grades': coffee + no_coffee,
    'Group': ['Coffee']*len(coffee) + ['No Coffee']*len(no_coffee)
})

# Run t-test
t_stat, p_val = ttest_ind(coffee, no_coffee, equal_var=False)
print(f"T-statistic: {t_stat:.2f}, P-value: {p_val:.4f}")

# Plot
plt.figure(figsize=(8, 5))

# Boxplot with hue and palette
sns.boxplot(x='Group', y='Grades', data=df, hue='Group', palette='coolwarm', dodge=False)

# Scatter the individual data points
sns.stripplot(x='Group', y='Grades', data=df, color='black', jitter=0.2, alpha=0.7)

# Title and labels
plt.title(f"Do Coffee-Drinkers Get Better Grades?\nT = {t_stat:.2f}, p = {p_val:.4f}")
plt.ylabel("Grades (from 0 to 100)")
plt.xlabel("")

# Don't even call legend() at all to avoid the warning
plt.tight_layout()
plt.show()
```
