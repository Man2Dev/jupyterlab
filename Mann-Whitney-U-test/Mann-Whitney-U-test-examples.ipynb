```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu

# Sample ordinal data: ratings from students
method_A = [5, 4, 3, 4, 2, 5, 4, 4, 3, 5]
method_B = [2, 2, 3, 1, 3, 2, 1, 2, 3, 2]

# Boxplot
sns.boxplot(data=[method_A, method_B])
plt.xticks([0, 1], ['Method A', 'Method B'])
plt.title("Student Ratings for Teaching Methods")
plt.ylabel("Rating (1=Bad, 5=Great)")
plt.show()

# Mann-Whitney U Test
stat, p = mannwhitneyu(method_A, method_B, alternative='two-sided')
print(f"Mann-Whitney U test: U={stat}, p-value={p}")
```

```python
# Small sample with an outlier
diet_1 = [3.2, 3.5, 15.0]  # One person had dramatic weight loss
diet_2 = [3.0, 2.8, 3.1]

# Plot
sns.boxplot(data=[diet_1, diet_2])
sns.stripplot(data=[diet_1, diet_2], color='black', jitter=True)
plt.xticks([0, 1], ['Diet 1', 'Diet 2'])
plt.title("Weight Loss Comparison (Small Sample with Outlier)")
plt.ylabel("Weight Loss (kg)")
plt.show()

# Mann-Whitney U Test
stat, p = mannwhitneyu(diet_1, diet_2, alternative='two-sided')
print(f"Mann-Whitney U test: U={stat}, p-value={p}")
```
